```python
sal = [15,22,24,25,30,23,34,45,19,20,52,21]
sal.sort()
sal
```

```python
#25% will take the average of first four numbers and then at their avg lets assume the average of first four number is 20.50 and then the 25% area will lie at the 20.50
```

# outliers:

```python
import pandas as pd 
ser = pd.Series(sal)
```

```python
ser
```

```python
#1st quantile 
q1 = ser.quantile(0.25)
print('Q1' , q1)

#3rd quantile 
q3 = ser.quantile(0.75)
print('Q3' , q3)

#find the iqr (inter quantile range) to identify outliers
iqr = q3-q1
print('IQR' , iqr)
```

# outlier detection formula

```python
upside = q3 + (1.5 * iqr)
print(upside )
```

```python
#anything above 46 is outlier.all are considered outliers
```

```python
lower_side = q1-(1.5 * iqr)
print(lower_side)
```

```python
#anything below 5.50 is outlier.
```

# pie chart

import all the same lib

```python
#load dataset from seaborn
mileage = sns.load_dataset('mpg')
mileage.head()
```

```python
avg_mpg = mileage.groupby(by='cylinders') [['mpg' , 'cylinders']].mean()
avg_mpg
```

```python
#pie
plt.pie(x = avg_mpg['mpg'],labels = avg_mpg['cylinders'])
plt.show
```

```python
#in the above pie chart we dont know which data iscovering more percentage.
```

```python
#for percentage 
plt.figure(figsize = (8,8))
plt.pie(x=avg_mpg['mpg'],labels = avg_mpg['cylinders'],explode = [0,0,0,0.05,0.3] ,autopct = '%.2f')
plt.title('avg_miles_per_cylinder_type')
plt.show()
```

```python
#we will start with the blue order is 3,4,5,6,8 and explode = distance which is 0 for the first three and then if we want to hihglight any part we have to enhance the distance 
```

```python
#autopct is for percentage %.2frepresent 2 digit after the point . and f represent 
```
